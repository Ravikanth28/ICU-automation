```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import pickle

data = pd.DataFrame({
    "hour": [1,2,3,4,5,6],
    "admissions": [2,3,5,6,8,10]
})

X = data[["hour"]]
y = data["admissions"]

model = RandomForestRegressor()
model.fit(X, y)

with open("model.pkl", "wb") as f:
    pickle.dump(model, f)
```
